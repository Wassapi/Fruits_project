Подготовим данные.
Все данные были разделены на тренировчную(70%) и тестовую выборку(30%).
Тренировочная выборка была разделена на тренировку (80%) и валидацию (20%). Подбор гиперпараметров будет определяться точностью на валидации.

```python
batch_size = 8

data_size = len(train_dataset)
validation_fraction = .2
test_fraction = 0.3

test_split = int(np.floor((test_fraction) * data_size))
val_split = int(np.floor((validation_fraction) * (data_size - test_split)))

ind = list(range(data_size))
np.random.seed(42)
np.random.shuffle(ind)

val_ind, train_ind, test_ind = ind[:val_split], ind[val_split: -test_split], ind[-test_split:]

train_sampler = SubsetRandomSampler(train_ind)
val_sampler = SubsetRandomSampler(val_ind)
test_sampler = SubsetSampler(test_ind)
all_sampler = SubsetRandomSampler(ind)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, 
                                           sampler=train_sampler)
val_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size,
                                         sampler=val_sampler)
test_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size,
                                         sampler=test_sampler)

all_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, 
                                           sampler=all_sampler)
```

Качество модели проверялось на тестовой выборке после окончания обучения. После того как модель и ее гиперпараметры были определены дообучение лучшей модели велось на всех данных.

```python
lr_base_params_list = [1e-4, 1e-5, 1e-6]   # Список со значениями learning_rate для базовых слоев нейронной сети.
lr_fc_params_list = [1e-2, 1e-3, 1e-4] # Список со значениями learning_rate для полносвязного слоя нейронной сети.

Hyperparams = namedtuple(
    "Hyperparams", 
    ['learning_rate_base_param', 'learning_rate_fc']
)  # Для записи значений базовых параметров на каждой итерации перебора комбинаций параметров.
RunResult = namedtuple(
    "RunResult", 
    ['model', 'train_history', 'val_history']
) # Для записи результатов тренировки на каждом этапе перебора комбинаций параметров после тренировки.
run_record = {} # Dict для записи "значений параметров" : "результат на валидации сети с этими параметрами".

for lr_base_params, lr_fc  in product(lr_base_params_list, lr_fc_params_list): # Перебираем комбинации параметров.
 
    model = models.resnet18(pretrained=True) # В основе моделей лежит натренированная модель Resnet18
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 16) # Заменяем полносвязный слой сети на новый с 16-ю выходами под наши классы.
    model = model.to(device) # Тренируем на GPU.
    
    loss = nn.CrossEntropyLoss()  # Определяем функцию потерь.
    optimizer = optim.SGD([{'params': model.fc.parameters(), 'lr': lr_fc},
                        {'params': [param for name, param in model.named_parameters() if 'fc' not in name]}], 
                          lr=lr_base_params, momentum=0.9) # Обучение ведется с разным learning rate.
    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, 3, eta_min=0, last_epoch=-1) # Также значение learning rate будет меняться от min к max.
    
    loss_history, train_history, val_history = train_model(
        model, train_loader, val_loader, loss, optimizer, scheduler, 10)

    params_set = {
        'lr_base_params': lr_base_params, 
        'lr_fc': lr_fc
         }
    print(params_set)
    run_record[Hyperparams(lr_base_params, lr_fc)
    ] = RunResult(model, train_history, val_history) # Запись результатов и значений параметров после тренировки.
```

```python
best_val_acc = None
best_hyperparams = None
best_model = None

for hyperparams, run_result in run_record.items():
    if best_val_acc is None or best_val_acc < run_result.val_history[-1]:
        best_val_acc = run_result.val_history[-1]
        best_hyperparams = hyperparams
        best_model = run_result.model
print("Best validation accuracy: %4.2f, best hyperparams: %s" % (best_val_acc, best_hyperparams))
```

```python
compute_accuracy(best_model, test_loader)
```

```python
model = best_model
model = model.to(device)
loss = nn.CrossEntropyLoss()
optimizer = optim.SGD([{'params': model.fc.parameters(), 'lr': best_hyperparams[1]},
                        {'params': [param for name, param in model.named_parameters() if 'fc' not in name]}], 
                      lr=best_hyperparams[0], momentum=0.9)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, 3, eta_min=0, last_epoch=-1)
loss_history, train_history, val_history = train_model(
    model, all_loader, val_loader, loss, optimizer, scheduler, 5)
print('Final val_accuracy: ', val_history[-1])
```
