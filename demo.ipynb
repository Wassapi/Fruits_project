```python
import numpy as np
import os
from PIL import Image
import torch
from torch import nn
from torchvision import models, utils
from torch.utils.data import Dataset, SubsetRandomSampler, Sampler
import matplotlib.pyplot as plt

from dataset import Fruits, fruit_list
from utils import visualize_samples, predicting_fruit

%matplotlib inline
device = torch.device("cuda:0")
```

```python
load_folder = "demonstration/"   # Folder for image loading.

classes_number = len(fruit_list)
```

```python
model = models.resnet18(pretrained=False)
num_filters = model.fc.in_features
model.fc = nn.Linear(num_filters, classes_number)
model.load_state_dict(torch.load('model_resnet18_comp.sh'))  # Loading of saved model weights.
model = model.to(device)
```

```python
# Showing of images and predicted class.
prediction = predicting_fruit(model, folder=load_folder, title_list=fruit_list)
```
